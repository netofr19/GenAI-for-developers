# Memory and Compute Optimizations

```python
import torch
```

```python
torch.finfo(torch.float32)
```

```python
pi = 3.14159265358979323846
pi_fp32 = torch.tensor(pi, dtype=torch.float32)
print(f"{pi_fp32.item():.20f}")
```

```python
def show_memory_consumption(tensor):
    memory_bytes = tensor.element_size() * tensor.numel()
    print(f"Memory consumption: {memory_bytes} bytes")
show_memory_consumption(pi_fp32)
```

```python
pi = 3.14159265358979323846
pi_fp16 = torch.tensor(pi, dtype=torch.float16)
print(f"{pi_fp16.item():.20f}")
```

```python
pi = 3.14159265358979323846
pi_bfp16 = torch.tensor(pi, dtype=torch.bfloat16)
print(f"{pi_bfp16.item():.20f}")
```

```python
pi = 3.14159265358979323846
pi_int8 = torch.tensor(pi, dtype=torch.int8)
print(pi_int8.item())
```
